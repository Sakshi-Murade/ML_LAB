```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("LossFromNetCrime_cleansed.csv").convert_dtypes()
```

```python
import matplotlib.pyplot as plt

# Drop missing values
data = data.dropna(subset=["COUNTRY", "2019_Complaints"])

# Convert data types
data["COUNTRY"] = data["COUNTRY"].astype(str)
data["2019_Complaints"] = pd.to_numeric(data["2019_Complaints"], errors="coerce")

# Sort data for better readability
data = data.sort_values(by="2019_Complaints", ascending=False)

# Create Bar Chart (Better for categorical data)
fig, ax = plt.subplots(figsize=(20, 10))
ax.bar(data["COUNTRY"], data["2019_Complaints"], color="skyblue")

# Labels and Title
ax.set_title("Number of Complaints per Country in 2019")
ax.set_xlabel("Country")
ax.set_ylabel("Number of Complaints")

# Rotate x-axis labels to avoid overlapping
plt.xticks(rotation=90)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("LossFromNetCrime.csv")

# Print first few rows
print(data.head())

# Drop NaN values
data = data.dropna()

# Convert numeric columns
numeric_cols = data.columns[1:]  # Exclude 'COUNTRY'
data[numeric_cols] = data[numeric_cols].apply(pd.to_numeric, errors="coerce")

# Verify Data
print(data.info())  # Check data types and missing values
```

```python
import pandas as pd
plt.figure(figsize=(21, 10))
plt.bar(data["Country"], data["2024_Complaints"], color="orange")
plt.title("Bar Chart of 2024 Complaints by Country")
plt.xlabel("Country")
plt.ylabel("Number of Complaints")
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.show()
```

```python
plt.figure(figsize=(21, 10))
plt.bar(data["Country"], data["2024_Complaints"], color="orange")
plt.title("Bar Chart of 2024 Complaints by Country")
plt.xlabel("Country")
plt.ylabel("Number of Complaints")
plt.xticks(rotation=90)  # Rotate for better visibility
plt.show()
```

```python

# Line Chart - Complaints Over the Years for Each Country
plt.figure(figsize=(12, 6))
for country in data["Country"].unique():
    subset = data[data["Country"] == country]
    plt.plot(["2019", "2020", "2021", "2022", "2023", "2024"], 
             subset.iloc[0, 1::2].values, marker="o", label=country)

plt.title("Complaints Over the Years by Country")
plt.xlabel("Year")
plt.ylabel("Number of Complaints")
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))  # Move legend outside
plt.show()
```

```python
# Scatter Plot - Complaints vs Losses in 2024
plt.figure(figsize=(8, 5))
plt.scatter(data["2024_Complaints"], data["2024_Losses"], color="red", alpha=0.6)
plt.title("Scatter Plot: Complaints vs Losses in 2024")
plt.xlabel("Number of Complaints")
plt.ylabel("Total Losses")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sort by complaints and take the top 10 (optional if too many countries)
top_data = data.nlargest(10, "2024_Complaints")

# Define explosion (separates slices slightly)
explode = [0.05] * len(top_data)  # Increase space between slices

plt.figure(figsize=(10, 8))
plt.pie(top_data["2024_Complaints"], labels=top_data["Country"], autopct="%1.1f%%", 
        startangle=90, explode=explode, pctdistance=0.65, wedgeprops={'edgecolor': 'black'})

# Add a circle to make it a donut chart (better visibility)
#centre_circle = plt.Circle((0, 0), 0.70, fc="white")
#plt.gca().add_artist(centre_circle)

plt.title("Complaint Distribution by Country (Top 10 - 2024)")
plt.show()
```

```python
# Histogram - Distribution of Complaints in 2024
plt.figure(figsize=(8, 5))
plt.hist(data["2024_Complaints"], bins=10, color="purple", alpha=0.7, edgecolor="black")
plt.title("Histogram of Complaints in 2024")
plt.xlabel("Number of Complaints")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
#Box Plot - Complaints Over the Years
plt.figure(figsize=(10, 6))
plt.boxplot([data["2019_Complaints"], data["2020_Complaints"], data["2021_Complaints"], 
             data["2022_Complaints"], data["2023_Complaints"], data["2024_Complaints"]], 
            labels=["2019", "2020", "2021", "2022", "2023", "2024"])

plt.title("Box Plot of Complaints Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Complaints")
plt.show()
```
